```python
import requests
import pandas as pd
import time

all_data = []
page = 1
per_page = 100  # pode ser até 200

while True:
    url = f"https://api.openbrewerydb.org/v1/breweries?page={page}&per_page={per_page}"
    response = requests.get(url)

    if response.status_code != 200:
        print(f"Erro ao acessar página {page}")
        break

    data = response.json()
    if not data:
        print("Todas as páginas foram lidas.")
        break

    all_data.extend(data)
    print(f"Página {page} lida com {len(data)} registros.")
    page += 1
    time.sleep(0.3)  # evita sobrecarga na API

# Transforma em DataFrame
df_all = pd.DataFrame(all_data)
print(f"\nTotal de registros coletados: {len(df_all)}")
df_all.head()
```

```python
from sqlalchemy import create_engine, text
import json

engine = create_engine('postgresql://dbadmin:dbadmin123@db:5432/fonte')

with engine.connect() as conn:
    for row in all_data:
        raw_json = json.dumps(row)
        conn.execute(text("INSERT INTO raw_breweries (raw_data) VALUES (:data)"), {"data": raw_json})
    conn.commit()

print(f"{len(all_data)} registros inseridos no banco com sucesso.")
```
